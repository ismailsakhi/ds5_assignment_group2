```python
#Ex1



import csv 

def load_csv(file_path: str):
    
    ''' 
    input : file path 
    output : records ( grades of each student as list (dict) ) 
    
    '''
    records = []
    with open(file_path, 'r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            records.append(row)
    return records


def calculate_avg(records):
    ''' 
    
    input : records 

    output : average grades by summing all grades and deviding them by number of records 
    
    '''
    total = sum(float(record['Grade']) for record in records)
    average = total / len(records)
    
    return average

 
def filter_records(records):
    
    ''' 
    
    input : records 

    output : a list of filtered  all grades above 80 in records 
    
    '''
    
    filtered_records = [record for record in records if float(record['Grade']) >= 80.0]
    
    return filtered_records

def print_report(filtered_records , average):
    
    '''
    input : calculated average and filtered records 
    
    output : report with average grade, name of the student and grade
    
    '''
    
    print(f"Average Grade: {average}")
    print("--------------------")
    print("Student Report")
    print("--------------")

    for record in filtered_records:
        print(f"Name: {record['Name']}")
        print(f"Grade: {record['Grade']}")
        print("--------------------")
        
def main():
    
    
    file_path = input(' enter csv path')
    records = load_csv(file_path)
    average = calculate_avg(records)
    
    print_report( records , average )
    
if __name__ == "__main__":
    main()
```
